```python
from bs4 import BeautifulSoup
import requests
```

```python
url ='https://www.numberfire.com/nba/daily-fantasy/daily-basketball-projections'
req = requests.get(url)
html = BeautifulSoup(req.content, 'html.parser')
```

```python
prices_html = html.find_all(class_="cost")
prices = [x.get_text(strip=True) for x in prices_html]
prices
```

```python
name_html = html.find_all(class_="full")
name = [x.get_text(strip=True) for x in name_html]
name
```

```python
opponent_html = html.find_all(class_="team-player__team ")
opponent = [x.get_text(strip=True) for x in opponent_html]
opponent
```
